```python
import pandas as pd
data={"字串":["台北","台中"],"int":[3000,1000],"float":[3.25,7.39]}

df=pd.DataFrame(data)
df
```

```python
#視覺美化資料表
#pip install jinja2
df.style.format(precision=2,thousands=",",decimal=".")\
.relabel_index(["列1","列2"])
```

```python
import numpy as np
data=np.random.rand(10,2)*5
df=pd.DataFrame(data,columns=["台北","台中"],index=pd.date_range(start="2023-09-01",periods=10))

df
```

```python
#這裡我沒聽到
#視覺美化資料表
def feel_condition(v):
    if v < 1.75:
        return "溫暖"
    elif v < 2.75:
        return "潮溼"
    else:
        return "日照"
    
df.style\
.format(feel_condition)\
.set_caption("體感溫度")\
.background_gradient(cmap="PuBuGn",vmin=1,vmax=5,axis=None)
```

```python
#隱藏不要讓別人看到
df=pd.DataFrame(np.random.rand(5,5))

df.style.hide(subset=[2],axis=0)\
    .hide(subset=[1,3],axis=1)
```

```python
df=pd.DataFrame(np.random.randn(10,4),columns=["a","b","c","d"])
display(df)

def style_negative( v, props=""):
    return props if v< 0 else None#None就不會有特效
df.style.map(style_negative,props="color:red;")
```
